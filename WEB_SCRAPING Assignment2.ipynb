**Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. This task will be done in following steps:**

**first get the webpage https://www.naukri.com/ Enter ‚ÄúData Analyst‚Äù in ‚ÄúSkill,Designations,Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field. Then click the search button. Then scrape the data for the first 10 jobs results you get. Finally create a dataframe of the scraped data.**

```python

driver=webdriver.Edge(r"D:\Greeshma\OneDrive - Infosys Limited\Desktop\msedgedriver.exe")
time.sleep(3)

#opening the website on automated browser page
url = "https://www.naukri.com/"
driver.get(url)

#entring desingnation and location as per question
desingnation=driver.find_element(By.CLASS_NAME,"suggestor-input")
desingnation.send_keys('Data Analyst')

#Entering Data into Search Bars :-
location=driver.find_element(By.XPATH,"//input[@class='suggestor-input 'and @placeholder='Enter location']")
location.send_keys("Bangalore")

#Clicking the search button:-
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()

#Creating empty lists for scraping data and use the scraped data to make DataFrame:-
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

#Scraping the job-titles:-
title=driver.find_elements(By.XPATH,"//a[@Class='title ellipsis']")
for i in title[0:10]:
    title=i.text
    job_title.append(title)
time.sleep(3)

#scraping the Location:-
location=driver.find_elements(By.XPATH,"//li[@Class='fleft br2 placeHolderLi location']//span[1]")
for i in location[0:10]:
    location=i.text
    job_location.append(location)
time.sleep(3)

#scraping the company_name:-
company=driver.find_elements(By.XPATH, '//a[@Class="subTitle ellipsis fleft"]')
for i in company[0:10]:
    company=i.text
    company_name.append(company)
time.sleep(3)


#scraping the experience_required:-
experience=driver.find_elements(By.XPATH,'//li[@Class="fleft br2 placeHolderLi experience"]/span[1]')
for i in experience[0:10]:
    experience=i.text
    experience_required.append(experience)
time.sleep(3)


# creating the dataframe from the scraped data and taking only first 10 jobs:-
JOB=pd.DataFrame({"job_title":job_title,"job_location":job_location,"company_name":company_name,"experience_required":experience_required})
JOB
```

**Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:**

After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data as usual.**

```python
driver=webdriver.Edge(r"D:\Greeshma\OneDrive - Infosys Limited\Desktop\msedgedriver.exe")
time.sleep(3)
Url="https://www.flipkart.com/"
driver.get(Url)
time.sleep(3)

close_popup=driver.find_element(By.XPATH,"//button[@class='_2KpZ6l _2doB4z']")
close_popup.click()

location=driver.find_element(By.XPATH,"//input[@Class='_3704LK']")
search=location.send_keys("sunglasses")
# clicking the search button:-
search_button=driver.find_element(By.CLASS_NAME, "_34RNph")
search_button.click()
time.sleep(10)

# creating empty lists for scraping data and use the scraped data to make DataFrame:-
Model_title = []
Brand_title = []
MRP_title =  []
Discount_title = []

start = 0
end = 3
for page in range(start,end):
    Model_Name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    time.sleep(3)
    for i in Model_Name:
        Model = i.text
        Model_title.append(Model)
    
    Brand_Name = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    time.sleep(3)
    for i in Brand_Name:
        Brand = i.text
        Brand_title.append(Brand)
    
    MRP = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    time.sleep(3)
    for i in MRP:
        MRP_Rs = i.text
        MRP_title.append(MRP_Rs)
    
    Discount = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    time.sleep(3)
    for i in Discount:
        discount = i.text
        Discount_title.append(discount)
        
    Next_Button = driver.find_element(By.XPATH,"//a//span[contains(text(),'Next')]")
    Next_Button.click()
    time.sleep(3)
    
print(len(Model_title),len(Brand_title),len(MRP_title),len(Discount_title))

df=pd.DataFrame({"Model":Model_title[:100],"Brand":Brand_title[:100],"MRP":MRP_title[:100],"Discount":Discount_title[:100]})
df

```

**Q8: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps: 1.First get the webpage https://www.azquotes.com/ 2. Click on Top Quotes 3. Than scrap a) Quote b) Author c) Type Of Quote**

```python
driver=webdriver.Edge(r"D:\Greeshma\OneDrive - Infosys Limited\Desktop\msedgedriver.exe")
time.sleep(3)
Url="https://www.azquotes.com/"
driver.get(Url)
time.sleep(3)
Topquotes=driver.find_element(By.XPATH,"//a[contains(text(),'Top Quotes')]")
Topquotes.click()

Quote = []
Author = []
Type_of_quote = []

# scraping filtered Quote from the given page

quo=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quo[0:1000]:
    ti=i.text
    Quote.append(ti)
    
# scraping Author from the given page

Aut=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in Aut[0:1000]:
    rat=i.text
    Author.append(rat)
    
# scraping Description from given page

desc=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in desc[0:1000]:
    de=i.text
    Type_of_quote.append(de)

    
FIL_DATA = pd.DataFrame({
    'Quote':Quote[0:1000],
    'Author':Author[0:1000],
    'Type_of_quote':Type_of_quote[0:1000]})

FIL_DATA
    
```
