**1.Write a python program which searches all the product under a particular product from www.amazon.in. The 
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
guitars.**

```python
driver=webdriver.Edge(r"D:\Greeshma\OneDrive - Infosys Limited\Desktop\msedgedriver.exe")
time.sleep(3)
#opening the url#
driver.get("https://www.amazon.in/")
driver.maximize_window()
WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.ID, "nav-search-submit-button")))
input1=input()
search_bar=driver.find_element(By.ID,'twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys(input1)
search_button=driver.find_element(By.ID,'nav-search-submit-button')
search_button.click()
time.sleep(5)

product_urls = []
start=0
end=3
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]') #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))  #appending the urls in product_urls list
    time.sleep(3)
    nxt_button=driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')#scraping the next button   
    nxt_button.click()
    time.sleep(2)
time.sleep(2)
len(product_urls)

#Make empty lists
Brand=[]
ProductName=[]
Price=[]
Availability=[]
ReturnExchange=[]
Delivery=[]

for url in product_urls:
    driver.get(url)
    time.sleep(3)
    
    try:
        brand=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
        
    try:
        product=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        ProductName.append(product.text)
    except NoSuchElementException:
        ProductName.append('-')
        
    try:
        cost=driver.find_element(By.XPATH,'//span[@class="a-offscreen"]')
        Price.append(cost.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        availability=driver.find_element(By.XPATH,'//div[@id="availability"]')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
    try:
        returnexchange=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]//div[2]//span')
        ReturnExchange.append(returnexchange.text)
    except NoSuchElementException:
        ReturnExchange.append('-')
        
    try:
        delivery=driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]//span[@class="a-text-bold"]')
        Delivery.append(delivery.text)
    except NoSuchElementException:
        Delivery.append('-')
        
print(len(Brand),len(ReturnExchange),len(Delivery),len(ProductName),len(Price),len(Availability))

df=pd.DataFrame({'Brand':Brand,'ProductName':ProductName,'ReturnExchange':ReturnExchange,'Price':Price,'Delivery':Delivery,'product_urls':product_urls,'Availability':Availability})
df
    



```

**3. Write a python program to access the search bar and search button on images.google.com and scrape 10 
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.**

```python
driver=webdriver.Edge(r"D:\Greeshma\OneDrive - Infosys Limited\Desktop\msedgedriver.exe")
time.sleep(3)

url='https://images.google.com/'
urls = []    # empty list for image urls to download
data = []    # empty list for storing data
search_item = ["Fruits", "Cars", "Machine Learning","Guitar","Cakes"]
for item in search_item:
    driver.get(url)  # performing a Google search for our selected items
    time.sleep(3)
    search_bar = driver.find_element(By.XPATH,"//textarea[@aria-label='Search']")     # finding search bar on Google page   
    search_bar.send_keys(item)      # send search item list 
    time.sleep(2)
    search_bar.submit() # now we will click on search button
    
    # scroll the page in order to generate more images on the website
    for _ in range(500):
        driver.execute_script("window.scrollBy(0,100)")        
        imgs = driver.find_elements(By.XPATH,"//img[@class='rg_i Q4LuWd']")
    img_url = []
    for image in imgs:
        source = image.get_attribute('src')
        if source is not None:
                if(source[0:4] == 'http'):
                    img_url.append(source)
    for i in img_url[:100]:      # getting only 100 images each
        urls.append(i)
for i in range(len(urls)):
    if i >= 300:
        break
    print("Obtaining {0} of {1} images".format(i, 300))
    response=requests.get(urls[i],verify=False)
    file = open(r"D:/Greeshma/OneDrive - Infosys Limited/Desktop/images/fliprobo image"+str(i)+".jpg", "wb")
    file.write(response.content)
```
